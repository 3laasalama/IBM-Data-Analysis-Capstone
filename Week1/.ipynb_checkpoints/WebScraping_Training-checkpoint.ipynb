```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://analytics.withgoogle.com/'
```

```python
data = requests.get(url).text
```

```python
gsite= BeautifulSoup(data, 'html5lib')
gsite
```

```python
for link in gsite.find_all('li'):
    print(link.get('class'))
```

```python
for image in gsite.find_all('img'):
    print(image.get('src'))
```

```python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html'
```

```python
data= requests.get(url).text
```

```python
table= BeautifulSoup(data, 'html5lib')
table
```

```python
for tr in table.find_all('tr'):
    row= tr.find_all('td')
    print('{} ---> {}'.format(row[2].getText(), row[3].getText()))
```
