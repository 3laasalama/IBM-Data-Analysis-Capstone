## Demo: How to work with database


Download database file.


```python
# We will copy and paste this url to download the file m4_survey_data.sqlite to use it later.
#!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data1.sqlite") # open a database connection
```

```python
conn
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

data = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
plt.figure(figsize=(8,6))
sns.distplot(a=df['ConvertedComp'], bins= 20, kde=True, hist= True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.figure(figsize=(8,6))
sns.boxplot(x='Age', data=df)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(8,6))
sns.scatterplot(x=df.Age, y=df.WorkWeekHrs, data=df)
plt.show()
```

```python
df.corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(8,6))
plt.scatter(x=df.WorkWeekHrs, y=df.CodeRevHrs, c='green', alpha=0.3 , s=df.Age*10)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
df2 = pd.read_sql_query('select * from DatabaseWorkedWith', conn)
```

```python
y= df2.groupby('DatabaseWorkedWith').count().sort_values(by='Respondent', ascending=False)
```

```python
y_5= y.head(5)
y_5
```

```python
plt.figure(figsize=(12,9))
plt.pie(y_5.Respondent, labels=['My SQL', 'Microsoft SQL Server', 'PostgreSQL', 'SQLite', 'MongoDB'], autopct='%.2f%%')
plt.title("DataBase Worked With Percentages")
plt.show()
```

```python
df3= pd.read_sql('select * from DatabaseDesireNextYear', conn)
df3
```

```python
df3_5= df3.groupby('DatabaseDesireNextYear').count().sort_values(by='Respondent', ascending=False)
```

```python
df3_5= df3_5.head(5)
df3_5
```

```python
plt.figure(figsize=(12,9))
plt.pie(df3_5.Respondent, labels=['PostgreSQL' , 'MongoDB', 'Redis', 'My SQL', 'Elasticsearch' ], autopct='%.2f%%')
plt.title("DataBase Worked With Percentages")
plt.show()
```

```python
lang_nexty = pd.read_sql_query('select * from LanguageWorkedWith', conn)
lang_nexty= lang_nexty.groupby('LanguageWorkedWith').count().sort_values(by='Respondent', ascending=False)
lang_nexty
```

```python
a= pd.read_sql_query('select * from WebFrameWorkedWith', conn)
a= a.groupby('WebFrameWorkedWith').count()
a
```

Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.
